```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
Heart_data = pd.read_csv('C:/Users/PARVEZ/Downloads/archive (1)/Heart_Disease_Prediction.csv')
```

```python
Heart_data['Heart Disease'] = Heart_data['Heart Disease'].map({"Absence" : 0 , "Presence": 1})
```

```python
Heart_data.describe()
```

```python
Heart_data['Heart Disease'].value_counts()
```

```python
Heart_data.groupby('Heart Disease').mean()
```

```python
Heart_data.isnull().sum()
```

```python
X = Heart_data.drop(columns = 'Heart Disease' , axis = 1)
Y = Heart_data['Heart Disease']
```

```python
X
```

```python
Y
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
standardized_data
```

```python
X = standardized_data
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, interaction_only=True)
X_poly = poly.fit_transform(X)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X_poly,Y , test_size = 0.2 , random_state = 2)
```

```python
X_train.shape , X_test.shape , X.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_classifier.fit(X_train, Y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print("\nClassification Report:")
print(classification_report(Y_test, y_pred))
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, y_pred))
```

```python
result_df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})
print(result_df)
```
